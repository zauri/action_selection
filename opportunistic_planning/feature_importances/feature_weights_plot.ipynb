```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
preds = pd.read_excel('data_tum_feature_importances.xlsx', header=0)
preds.set_index('Feature', drop=True, inplace=True)
```

```python
ax1 = plt.boxplot(preds, meanline=True, showmeans=True)
#ax1 = sns.swarmplot(data=preds2)
plt.xticks([1, 2, 3, 4, 5], preds.index)
plt.ylabel('Feature weight', fontweight='bold', labelpad=10)
plt.xlabel('Feature', fontweight='bold', labelpad=10)
plt.show()
```

```python
preds2 = preds.T
```

```python
#sns.set_style('whitegrid')
ax = sns.boxplot(data=preds2, color='lightgrey', showmeans=True, 
                 meanline=True, linewidth=0.5,)

ax.axes.set_title('Percentage of feature weights (TUM data)', 
                  fontweight='bold', pad=10)
ax.set_xlabel('Feature', fontsize=10, fontweight='bold')
ax.set_ylabel('Feature weight', fontsize=10, fontweight='bold')
ax = sns.swarmplot(data=preds2, color='Red')
plt.show()
```
