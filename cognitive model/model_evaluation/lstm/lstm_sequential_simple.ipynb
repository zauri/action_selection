```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers.core import Dense
```

```python
train_file = 'all_data_train.csv'
```

```python
train = pd.read_csv(train_file, header=0, skiprows=98)
data = []
target = []

for index, row in train.iterrows():
    length = np.count_nonzero(~pd.isnull(row))
    data.append(row.values[:1].tolist())
    target.append(row.values[1:3])
```

```python
data_num = []

for array in data:
    for value in array:
        value = ord(value)
        data_num.append(value)
```

```python
target_num = []

for array in target:
    values = [ord(x) for x in array]
    target_num.append(values)
```

```python
X = np.array(data_num).reshape(len(data_num),1,1)
Y = np.array(target_num, dtype='object')
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(1,1)))
model.add(Dense(2))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, Y, epochs=500, validation_split=0.2, batch_size=3)
```

```python
test_input = data_num[:1]
test_input = np.array(test_input)
test_input = test_input.reshape((1,1,1))
test_output = model.predict(test_input, verbose=0)
```
