```python
df = pd.read_csv('data/single_step_df_ints_2022-06-03_encoded.csv', header=0)
```

```python
no_seen_items = df.copy()
```

```python
no_seen_items.to_csv('data/single_step_df_no_seen_items_2022-08-16.csv', header=True, index=False)
```
