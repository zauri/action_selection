```python
import pandas as pd
import math
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('data.txt', skiprows=3, names=['pos1','pos2','pos3','pos4',
                                                 'pos5','pos6','pos7','pos8'])
```

```python
df.index += 1
```

```python
df
```

```python
for row in range(1,len(df)+1):
    if pd.isna(df.loc[row,'pos2']):
        df.loc[row-1,'start'] = df.loc[row,'pos1']
        df = df.drop(row)
    elif row % 3 == 0:
        pos1 = df.loc[row,'pos1']
        pos2 = df.loc[row,'pos2']
        pos3 = df.loc[row,'pos3']
        pos4 = df.loc[row,'pos4']
        pos5 = df.loc[row,'pos5']
        pos6 = df.loc[row,'pos6']
        pos7 = df.loc[row,'pos7']
        pos8 = df.loc[row,'pos8']
        df.loc[row-2,'target'] = f"{pos1, pos2, pos3, pos4, pos5, pos6, pos7, pos8}"
        df = df.drop(row)

df.reset_index(inplace=True, drop=True)
```

```python
df.to_csv('h03_data.csv', header=True)
```
