```python
import pandas as pd
```

```python
df = pd.read_csv('data/task_environments_2022-06-01_encoded_sequences.csv', header=0)
```

```python
values = df['encoded_sequence'].value_counts().keys().tolist()
counts = df['encoded_sequence'].value_counts().tolist()
```

```python
occurrences = {value: count for value, count in zip(values, counts) if int(count) > 1}
```

```python
occurrences
```

```python
df = df.loc[df['encoded_sequence'].isin(occurrences.keys())].reset_index()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```
