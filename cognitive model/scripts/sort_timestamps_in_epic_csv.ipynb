```python
df = pd.read_csv('data/EPIC_100_annotated_episodes_amended.csv', header=0, index_col=False)
```

```python
df
```

```python
df_new = pd.DataFrame()
```

```python
for video_id in df['video_id'].unique():
    id_df = df.loc[df['video_id'] == video_id]
    id_df.sort_values('start_timestamp', inplace=True)
    df_new = df_new.append(id_df)
```

```python
df_new
```

```python
df_new.to_csv('data/EPIC_100_annotated_episodes_amended2.csv', index=False, header=True)
```
